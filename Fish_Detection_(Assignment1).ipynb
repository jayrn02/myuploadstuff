# Training Notebook



*   **yolov8n.pt (Nano)**: Fastest, lowest VRAM, least accurate.
*   **yolov8s.pt (Small)**: Good balance (what you likely used).
*   **yolov8m.pt (Medium**): More accurate, slower, needs more VRAM.
*   **yolov8l.pt (Large)**: Even more accurate, slower, needs more VRAM.
*   **yolov8x.pt (Extra Large)**: Potentially most accurate, slowest, needs significant



- **epochs / patience**:

- **lr0 (Initial Learning Rate)**:  
  Controls how big the adjustment steps are during training. The default is often good, but sometimes slightly lowering it (e.g., `lr0=0.005` instead of default `0.01`) can help if validation loss is unstable.

- **batch (Batch Size)**:  
  Number of images processed at once. Limited by GPU memory in Colab. Larger batches can sometimes stabilize training but require more VRAM. Default is often 16. Try 8 or 32 if memory allows/errors occur.

- **imgsz (Image Size)**:  
  Train and infer at the same size (e.g., `imgsz=640`). Larger sizes (e.g., `1280`) can detect smaller objects better but require much more VRAM and time. Smaller sizes (e.g., `320`) are faster but may miss details.

```python
import supervision as sv
import cv2
from ultralytics import YOLO
import numpy as np
from collections import defaultdict

model = YOLO('/content/datasets/runs/detect/train/weights/best.pt') # Use best model's path
tracker = sv.ByteTrack()


model_path = '/content/datasets/runs/detect/train/weights/best.pt' # Use your actual best model path
video_path = "/content/Fish_Stable_1.mp4" # Or "/content/fish_moving_2.mp4"
output_video_path_counting = "/content/output_fish_counting.mp4"
```

Detecting and Counting

```python
# BoxAnnotator draws the boxes
box_annotator = sv.BoxAnnotator(
    thickness=2 # Control box line thickness
)

# LabelAnnotator handles drawing the text labels
label_annotator = sv.LabelAnnotator(
    text_thickness=1, # Control text line thickness
    text_scale=0.6,    # Control font size scale
    text_padding=2,    # Padding around text
    text_position=sv.Position.TOP_CENTER # Where to place the label relative to the box
)
# TraceAnnotator draws the tracking paths
trace_annotator = sv.TraceAnnotator(
    trace_length=30,
    thickness=2,
    position=sv.Position.CENTER # Anchor point for the trace relative to the box
)
# *** End Correction ***

# Dictionary to store track history (track_id -> list of center points)
track_history = defaultdict(lambda: [])
# Dictionary to store calculated speeds (track_id -> speed)
track_speeds = defaultdict(lambda: 0.0)

# Video setup
video_path = "/content/Fish_Stable_1.mp4"

# video_path = "/content/fish_moving_2.mp4"

cap = cv2.VideoCapture(video_path)
if not cap.isOpened():
    print(f"Error: Could not open video file: {video_path}")
    exit()

fps = cap.get(cv2.CAP_PROP_FPS)
if fps <= 0: # Check if fps is valid
    print("Warning: Could not get valid FPS from video. Assuming 30 FPS.")
    fps = 30
dt = 1.0 / fps # Time interval between frames

# Output video setup
frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
output_video_path = "/content/output_fish_tracking_speed.mp4"
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
out = cv2.VideoWriter(output_video_path, fourcc, fps, (frame_width, frame_height))
print(f"Processing video... FPS: {fps:.2f}, dt: {dt:.4f}s")
# --- End Initialization ---

# --- Video Loop (Needs slight change to use both annotators) ---
frame_count = 0
while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        print("End of video reached or read error.")
        break

    frame_count += 1
    # 1. Detect
    results = model.predict(frame, verbose=False, conf=0.3)

    # 2. Convert detections to supervision format
    detections = sv.Detections.from_ultralytics(results[0])

    # 3. Track
    detections = tracker.update_with_detections(detections=detections)

    # Prepare labels and calculate speed
    labels = []
    for i in range(len(detections)):
        xyxy = detections.xyxy[i]
        confidence = detections.confidence[i]
        class_id = detections.class_id[i]
        tracker_id = detections.tracker_id[i]

        center_x = int((xyxy[0] + xyxy[2]) / 2)
        center_y = int((xyxy[1] + xyxy[3]) / 2)
        center_point = (center_x, center_y)

        history = track_history[tracker_id]
        history.append(center_point)
        if len(history) > 2:
            history.pop(0)

        speed_pixels_per_second = 0
        if len(history) == 2:
            prev_point = history[0]
            curr_point = history[1]
            distance_pixels = np.linalg.norm(np.array(curr_point) - np.array(prev_point))
            speed_pixels_per_second = distance_pixels / dt
            track_speeds[tracker_id] = speed_pixels_per_second

        current_speed = track_speeds[tracker_id]
        # label = f"ID:{tracker_id} Spd:{current_speed:.1f}" # More detailed label if needed
        label = f"ID:{tracker_id}" # Simpler label
        labels.append(label)


    # --- Annotation using Supervision Annotators ---
    annotated_frame = frame.copy()

    # 1. Draw the boxes first
    annotated_frame = box_annotator.annotate(
        scene=annotated_frame,
        detections=detections
    )
    # 2. Draw the labels using the generated labels list
    annotated_frame = label_annotator.annotate(
        scene=annotated_frame,
        detections=detections,
        labels=labels
    )
    # 3. Draw the tracking trails
    annotated_frame = trace_annotator.annotate(
        scene=annotated_frame,
        detections=detections
    )

    # Add overall frame count and fish count
    num_fish_in_frame = len(detections)
    cv2.putText(annotated_frame, f"Frame: {frame_count}", (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 0.8, (255, 255, 255), 2)
    cv2.putText(annotated_frame, f"Fish Count: {num_fish_in_frame}", (10, 60), cv2.FONT_HERSHEY_SIMPLEX, 0.8, (0, 255, 0), 2)

    # Add individual speeds next to boxes
    for i in range(len(detections)):
         if detections.tracker_id[i] in track_speeds:
             xyxy = detections.xyxy[i]
             tracker_id = detections.tracker_id[i]
             speed = track_speeds[tracker_id]
             pos = (int(xyxy[0]), int(xyxy[1] - 20))
             cv2.putText(annotated_frame, f"Spd:{speed:.1f}", pos, cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 255), 2)


    out.write(annotated_frame)


cap.release()
out.release()
print(f"Finished processing. Output video saved to: {output_video_path}")
```

```python
# --- Block 1: Fish Detection and Counting ---


print("--- Starting Block 1: Detection and Counting ---")

# --- Initialization ---
model_path = '/content/datasets/runs/detect/train/weights/best.pt' # Use your actual best model path
video_path = "/content/Fish_Stable_1.mp4" # Or "/content/fish_moving_2.mp4"
output_video_path_counting = "/content/output_fish_counting.mp4"

model = YOLO(model_path)
tracker = sv.ByteTrack()

# Annotators for boxes and basic ID labels
box_annotator = sv.BoxAnnotator(thickness=2)
label_annotator = sv.LabelAnnotator(
    text_thickness=1,
    text_scale=0.6,
    text_padding=2,
    text_position=sv.Position.TOP_CENTER
)

# Video setup
cap = cv2.VideoCapture(video_path)
if not cap.isOpened():
    print(f"Error: Could not open video file: {video_path}")
    # exit() # Avoid exiting if running in a notebook, just stop this block
else:
    fps = cap.get(cv2.CAP_PROP_FPS)
    if fps <= 0:
        print("Warning: Could not get valid FPS from video. Assuming 30 FPS.")
        fps = 30

    # Output video setup for counting
    frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
    fourcc = cv2.VideoWriter_fourcc(*'mp4v')
    out_counting = cv2.VideoWriter(output_video_path_counting, fourcc, fps, (frame_width, frame_height))
    print(f"Processing video for counting... Output: {output_video_path_counting}")

    # --- Video Loop for Counting ---
    frame_count = 0
    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            print("End of video reached or read error (Counting Block).")
            break

        frame_count += 1

        # 1. Detect
        results = model.predict(frame, verbose=False, conf=0.3) # Adjust confidence if needed

        # 2. Convert detections
        detections = sv.Detections.from_ultralytics(results[0])

        # 3. Track
        detections = tracker.update_with_detections(detections=detections)

        # 4. Prepare simple labels (just IDs)
        labels = []
        for i in range(len(detections)):
            tracker_id = detections.tracker_id[i]
            label = f"ID:{tracker_id}"
            labels.append(label)

        # 5. Annotate Frame for Counting
        annotated_frame = frame.copy()

        # Draw boxes
        annotated_frame = box_annotator.annotate(
            scene=annotated_frame,
            detections=detections
        )
        # Draw ID labels
        annotated_frame = label_annotator.annotate(
            scene=annotated_frame,
            detections=detections,
            labels=labels
        )

        # Add overall frame count and fish count
        num_fish_in_frame = len(detections)
        cv2.putText(annotated_frame, f"Frame: {frame_count}", (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 0.8, (255, 255, 255), 2)
        cv2.putText(annotated_frame, f"Fish Count: {num_fish_in_frame}", (10, 60), cv2.FONT_HERSHEY_SIMPLEX, 0.8, (0, 255, 0), 2)

        # 6. Write frame to output counting video
        out_counting.write(annotated_frame)

    # --- Cleanup for Counting ---
    cap.release()
    out_counting.release()
    print(f"Finished counting processing. Output video saved to: {output_video_path_counting}")

print("--- Finished Block 1 ---")
```

```python
# --- Block 2: Fish Speed Detection ---

import supervision as sv
import cv2
from ultralytics import YOLO
import numpy as np
from collections import defaultdict

print("--- Starting Block 2: Speed Detection ---")

# --- Initialization (Needs to be repeated) ---
model_path = '/content/datasets/runs/detect/train/weights/best.pt' # Use your actual best model path
video_path = "/content/Fish_Stable_1.mp4" # Or "/content/fish_moving_2.mp4" - USE THE SAME INPUT VIDEO
output_video_path_speed = "/content/output_fish_speed.mp4"

model = YOLO(model_path)
tracker = sv.ByteTrack()

# Annotators for boxes, IDs, and trails
box_annotator = sv.BoxAnnotator(thickness=2)
label_annotator = sv.LabelAnnotator(
    text_thickness=1,
    text_scale=0.6,
    text_padding=2,
    text_position=sv.Position.TOP_CENTER
)
trace_annotator = sv.TraceAnnotator(
    trace_length=30, # How long the trail is
    thickness=2,
    position=sv.Position.CENTER
)

# Dictionaries for speed calculation
track_history = defaultdict(lambda: [])
track_speeds = defaultdict(lambda: 0.0)

# Video setup (Needs to be repeated)
cap = cv2.VideoCapture(video_path)
if not cap.isOpened():
    print(f"Error: Could not open video file: {video_path}")
    # exit() # Avoid exiting if running in a notebook, just stop this block
else:
    fps = cap.get(cv2.CAP_PROP_FPS)
    if fps <= 0:
        print("Warning: Could not get valid FPS from video. Assuming 30 FPS.")
        fps = 30
    dt = 1.0 / fps # Time interval between frames - CRITICAL for speed

    # Output video setup for speed
    frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
    fourcc = cv2.VideoWriter_fourcc(*'mp4v')
    out_speed = cv2.VideoWriter(output_video_path_speed, fourcc, fps, (frame_width, frame_height))
    print(f"Processing video for speed... FPS: {fps:.2f}, dt: {dt:.4f}s. Output: {output_video_path_speed}")

    # --- Video Loop for Speed ---
    frame_count = 0
    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            print("End of video reached or read error (Speed Block).")
            break

        frame_count += 1

        # 1. Detect
        results = model.predict(frame, verbose=False, conf=0.3) # Adjust confidence if needed

        # 2. Convert detections
        detections = sv.Detections.from_ultralytics(results[0])

        # 3. Track
        detections = tracker.update_with_detections(detections=detections)

        # 4. Calculate Speed & Prepare Labels
        labels = [] # Simple labels for the label_annotator
        current_frame_speeds = {} # Store speeds calculated in this frame

        for i in range(len(detections)):
            xyxy = detections.xyxy[i]
            tracker_id = detections.tracker_id[i]

            # Calculate center point for history/speed
            center_x = int((xyxy[0] + xyxy[2]) / 2)
            center_y = int((xyxy[1] + xyxy[3]) / 2)
            center_point = (center_x, center_y)

            # Update history
            history = track_history[tracker_id]
            history.append(center_point)
            # Keep history length limited (e.g., last 2 points needed for speed)
            if len(history) > 2: # Only need previous and current
                history.pop(0)

            # Calculate speed
            speed_pixels_per_second = 0.0 # Default speed if not enough history
            if len(history) == 2:
                prev_point = history[0]
                curr_point = history[1]
                # Calculate Euclidean distance
                distance_pixels = np.linalg.norm(np.array(curr_point) - np.array(prev_point))
                # Speed = distance / time
                speed_pixels_per_second = distance_pixels / dt
                track_speeds[tracker_id] = speed_pixels_per_second # Update stored speed

            # Get the most recently calculated/stored speed for this ID
            current_speed = track_speeds[tracker_id]
            current_frame_speeds[tracker_id] = current_speed # Store for drawing later

            # Create simple label for label_annotator
            label = f"ID:{tracker_id}"
            labels.append(label)


        # 5. Annotate Frame for Speed
        annotated_frame = frame.copy()

        # Draw boxes
        annotated_frame = box_annotator.annotate(
            scene=annotated_frame,
            detections=detections
        )
        # Draw ID labels
        annotated_frame = label_annotator.annotate(
            scene=annotated_frame,
            detections=detections,
            labels=labels
        )
        # Draw tracking trails
        annotated_frame = trace_annotator.annotate(
            scene=annotated_frame,
            detections=detections
        )

        # Add overall frame count (optional, but good context)
        cv2.putText(annotated_frame, f"Frame: {frame_count}", (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 0.8, (255, 255, 255), 2)

        # Add individual speeds next to boxes using cv2.putText
        for i in range(len(detections)):
            tracker_id = detections.tracker_id[i]
            if tracker_id in current_frame_speeds: # Check if speed was calculated
                 xyxy = detections.xyxy[i]
                 speed = current_frame_speeds[tracker_id]
                 # Position speed text slightly offset (e.g., below the ID label)
                 pos = (int(xyxy[0]), int(xyxy[1] - 5)) # Adjust position as needed
                 cv2.putText(annotated_frame, f"Spd:{speed:.1f}", pos, cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 255), 2) # Yellow text

        # 6. Write frame to output speed video
        out_speed.write(annotated_frame)

    # --- Cleanup for Speed ---
    cap.release()
    out_speed.release()
    print(f"Finished speed processing. Output video saved to: {output_video_path_speed}")

print("--- Finished Block 2 ---")
```
